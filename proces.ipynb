```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

#####################################
c1 = 1
c2 = 12
#####################################
data_rows_1 = 110
data_rows_2 = 80
data_rows_3 = 110
data_rows_4 = 110
data_rows_5 = 110
#####################################
data_rows_s_1 = 15
data_rows_s_2 = 15
data_rows_s_3 = 15
data_rows_s_4 = 15
data_rows_s_5 = 55
#####################################
#dictionary of data rows
data_rows_end = {1: data_rows_1, 2: data_rows_2, 3: data_rows_3, 4: data_rows_4, 5: data_rows_5}
data_row_start = {1: data_rows_s_1, 2: data_rows_s_2, 3: data_rows_s_3, 4: data_rows_s_4, 5: data_rows_s_5}

#####################################
#area of the specimen in a dictionary 
b_1 = 13.23
h_1 = 3.07
b_2 = 12.88
h_2 = 3.13
b_3 = 3.12
h_3 = 13.21
b_4 = 3.1
h_4 = 13.16
b_5 = 3.12
h_5 = 12.78

area = {1: b_1 * h_1, 2: b_2 * h_2, 3: b_3 * h_3, 4: b_4 * h_4, 5: b_5 * h_5}
#####################################


plt.figure()
for k in np.arange(1, 6, 1):

    # Read the CSV file
    df_strain = pd.read_csv('compression/' + str(k) + '_export.csv', skiprows=1)
    df_stress = pd.read_csv('compression/' + str(k) + '_export.csv', skiprows=1)

    # Slice the DataFrame to get the required columns and rows
    extension = df_strain.iloc[data_row_start[k]:data_rows_end[k], c1]
    load = df_stress.iloc[data_row_start[k]:data_rows_end[k], c2]
    
    # Ensure lengths match before performing calculations
    if len(extension) != len(load):
        print(f'Error: Length mismatch for specimen {k}')
        continue

    #stress calculation
    stress = (-1) * load / area[k]

    #strain calculation
    strain = (-1) * extension 

    # Perform linear regression
    slope, intercept, r, p, std_err = stats.linregress(strain, stress)
    
    slope = round(slope, 1)
    intercept = round(intercept, 1)
    # Print the slope and intercept
    print(f'Specimen {k} - Slope: {slope}, Intercept: {intercept}')

    # Plotting
    plt.scatter(strain, stress, label=f'{k}: S[MPa] = {intercept}[MPa] + {slope}[MPa] strain ', s=1)    
    # Plot the linear regression line
    plt.plot(strain, intercept + slope * strain, linestyle='--')
 

plt.title("Compression test")
plt.grid()
plt.xlabel('Strain (mm/mm)')
plt.ylabel('Stress (MPa)')
plt.legend(loc=4,fontsize=8)
plt.savefig('Specimen_RawData_All.png')
plt.show()
```
